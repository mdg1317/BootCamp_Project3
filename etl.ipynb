```python
# Import dependencies 
import pandas as pd
```

```python
# Read CSV into DataFrame
ufo_df = pd.read_csv("static/resources/ufo_orig.csv", on_bad_lines="skip")
ufo_df.head()
```

```python
# Drop rows with empty columns
ufo_df = ufo_df.dropna()
ufo_df.head()
```

```python
# Drop data from outside US
ufo_df = ufo_df[ufo_df.country == "us"]
ufo_df.head()
```

```python
# Rename/delete columns
ufo_df = ufo_df.drop(columns=["country", "date posted", "duration (hours/min)"])
ufo_df = ufo_df.rename(columns={"duration (seconds)": "duration"})
ufo_df.head()
```

```python
# Reorder columns
ufo_df = ufo_df[["datetime", "latitude", "longitude", "city", "state", "duration", "shape", "comments"]]
ufo_df
```

```python
# Check column types
ufo_df.info()
```

```python
# Change column types
ufo_df = ufo_df.astype({"latitude": "float64", "duration": "float64"})
ufo_df = ufo_df.astype({"duration": "int32"})
ufo_df["datetime"] = pd.to_datetime(ufo_df["datetime"], errors="coerce")
ufo_df.info()
```

```python
# Write new CSV to file
ufo_df.to_csv("static/resources/ufo.csv")
```
